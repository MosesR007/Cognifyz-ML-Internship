```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('Dataset .csv')
df = df[[
    "Cuisines",
    "City",
    "Country Code",
    "Price range",
    "Has Online delivery",
    "Has Table booking",
    "Aggregate rating"
]]
```

```python
df = df.dropna(subset=["Cuisines"])

df["City"].fillna("Unknown", inplace=True)
df["Aggregate rating"].fillna(df["Aggregate rating"].median(), inplace=True)
```

```python
city_encoder = LabelEncoder()
delivery_encoder = LabelEncoder()
table_encoder = LabelEncoder()

df["City"] = city_encoder.fit_transform(df["City"])
df["Has Online delivery"] = delivery_encoder.fit_transform(df["Has Online delivery"])
df["Has Table booking"] = table_encoder.fit_transform(df["Has Table booking"])
```

```python
cuisine_encoder = LabelEncoder()
df["Cuisine_Label"] = cuisine_encoder.fit_transform(df["Cuisines"])
```

```python
X = df.drop(["Cuisines", "Cuisine_Label"], axis=1)
y = df["Cuisine_Label"]
```

```python
counts = y.value_counts()
valid_classes = counts[counts >= 2].index

df = df[df["Cuisine_Label"].isin(valid_classes)]

X = df.drop(["Cuisines", "Cuisine_Label"], axis=1)
y = df["Cuisine_Label"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)
```

```python
model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)

model.fit(X_train, y_train)
print("Model trained successfully")
```

```python
y_pred = model.predict(X_test)

print("Accuracy :", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred, average="weighted"))
print("Recall   :", recall_score(y_test, y_pred, average="weighted"))

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

```python
sample_X = X_test.iloc[[0]]
sample_y = y_test.iloc[0]

pred = model.predict(sample_X)[0]

print("Predicted Cuisine:", cuisine_encoder.inverse_transform([pred])[0])
print("Actual Cuisine   :", cuisine_encoder.inverse_transform([sample_y])[0])
```

```python
sample_X = X_test.iloc[[0]]
sample_y = y_test.iloc[0]

pred_label = model.predict(sample_X)[0]

predicted_cuisine = cuisine_encoder.inverse_transform([pred_label])[0]
actual_cuisine = cuisine_encoder.inverse_transform([sample_y])[0]

print("Predicted Cuisine:", predicted_cuisine)
print("Actual Cuisine   :", actual_cuisine)
```
